```julia
function strong_kron(A,B ::Array{T,N}) where {T<:Number,N}
    @assert size(A,3) == size(B,1)
    α=size(A,1)
    μ=size(A,2)
    β=size(A,3)
    ρ=size(B,2)
    γ=size(B,3)
    
    A = reshape(A,α*μ,β)
    B = reshape(B,β,ρ*γ)
    
    C = A*B
    
    C=reshape(C,α,μ*ρ,γ)
    
    return C
end
nothing;
```

```julia
function evaluate(S,n)
    N = length(S)
    A = copy(S[1])
    for j ∈ 2:N
        A=strong_kron(A,copy(S[j]))
    end
    A = reshape(A,n...)
    return A
end
```

```julia
function trig(L,a,b,c)
    U_0 = zeros(1,1,2)
    U_0[1,1,1] = a*cos(π*c) + b*sin(π*c)
    U_0[1,1,2] = b*cos(π*c) - a*sin(π*c)
    
    U_decomp = [U_0]
    temp = π*c
    for i = 1:L-1
        temp = temp/2   #2^-l pi c
        U_l = zeros(2,2,2)
        U_l[1,1,1] = cos(temp)
        U_l[1,1,2] = sin(temp)
        
        U_l[1,2,1] = cos(temp)
        U_l[1,2,2] = -sin(temp)
        
        U_l[2,1,1] = -sin(temp)
        U_l[2,1,2] = cos(temp)
        
        U_l[2,2,1] = sin(temp)
        U_l[2,2,2] = cos(temp)
        
        push!(U_decomp,U_l)
    end
    U_L = zeros(2,2,2)
    
    U_L[1,1,1] = (cos(temp*0)+cos(temp*(-1)))/2  #(t-1)/2
    U_L[1,1,2] = (cos(temp*0)-cos(temp*(-1)))/2 
    
    
    U_L[1,2,1] = (cos(temp*1)+cos(temp*(0)))/2  #(t+1)/2
    U_L[1,2,2] = (cos(temp*1)-cos(temp*(0)))/2
    
    U_L[2,1,1] = (sin(temp*0)+sin(temp*(-1)))/2 #(t-1)/2
    U_L[2,1,2] = (sin(temp*0)-sin(temp*(-1)))/2
    
    U_L[2,2,1] = (sin(temp*1)+sin(temp*(0)))/2  #(t+1)/2
    U_L[2,2,2] = (sin(temp*1)-sin(temp*(0)))/2
   
    push!(U_decomp,U_L)
    
    U_last = ones(2,1,1)
    push!(U_decomp,U_last)
    return U_decomp
end
        
```
